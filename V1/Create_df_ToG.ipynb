```python
puntos = {
    "Tipo": ("puntos", "bronce", "plata", "oro", "morado", "rojo", "amarillo"),
    "Puntos_Actuales": (0, 0, 0, 0, 0, 0, 0),
    "Puntos_Maximos" : (10000, 10000, 10000, 10000, 200, 200, 200)
}
```

```python
import pandas as pd

df = pd.DataFrame(puntos).set_index("Tipo")
df.to_json("puntos.json")
df
```

```python
df["Puntos_Restantes"] = df.Puntos_Maximos - df.Puntos_Actuales
df.to_json("puntos.json")
df
```
