```python
import pandas as pd
```

```python
def read_df(path = "puntos.json"):
    try:
        df = pd.read_json(path)
        return df
    except:
        print("No se ha encontrado el archivo")
```

```python
def print_info(tickets : int, df : pd.DataFrame):
    
    print(df.head())

    print(f"\nPlata = {df.Puntos_Actuales['plata']}/{df.Puntos_Maximos['plata']}({df.Puntos_Restantes['plata']})\
        oro = {df.Puntos_Actuales['oro']}/{df.Puntos_Maximos['oro']}({df.Puntos_Restantes['oro']})\
        morado = {df.Puntos_Actuales['morado']}/{df.Puntos_Maximos['morado']}({df.Puntos_Restantes['morado']}))\
        rojo = {df.Puntos_Actuales['rojo']}/{df.Puntos_Maximos['rojo']}({df.Puntos_Restantes['rojo']}))\
        \n")

    print(f"\n{tickets} disponibles\n")

    rondas = int(tickets/10)
    enemigos = 3 * rondas
    bronce = int(60 * 1.1) * rondas
    plata = int(20 * 1.2) * rondas
    puntos = int(30 * 1.05) * rondas
    print(f"Fase 1 {rondas} rondas: \n Estatuas: {bronce} bronce y {plata} plata \n {puntos} puntos \n {enemigos} enemigos morado\n")

    plata = int(50 * 1.2) * rondas
    oro = 20 * rondas
    print(f"Fase 2 {rondas} rondas: \n Estatuas: {plata} plata y {oro} oro \n {puntos} puntos \n {enemigos} enemigos rojo\n")

    bronce = int(20 * 1.1) * rondas
    oro = 50 * rondas
    print(f"Fase 3 {rondas} rondas: \n Estatuas: {bronce} bronce y {oro} oro \n {puntos} puntos \n {enemigos} enemigos amarillo\n")

    rondas = int(tickets/12)
    enemigos = 5 * rondas
    bronce = int(160 * 1.1) * rondas
    puntos = int(40 * 1.05) * rondas
    print(f"Fase 4 {rondas} rondas: \n Estatuas: {bronce} bronce \n {puntos} puntos \n {enemigos} enemigos rojo\n")

    enemigos = 3 * rondas
    plata = int(130 * 1.2) * rondas
    print(f"Fase 5 {rondas} rondas: \n Estatuas: {plata} plata \n {puntos} puntos \n {enemigos} enemigos morado\n")

    oro = 110 * rondas
    print(f"Fase 6 {rondas} rondas: \n Estatuas: {oro} oro \n {puntos} puntos \n {enemigos} enemigos rojo\n")

    bronce = int(20 * 1.1) * rondas
    plata = int(20 * 1.2) * rondas
    oro = 20 * rondas
    puntos = int(140 * 1.05) * rondas
    print(f"Fase 7 {rondas} rondas: \n Estatuas: {bronce} bronce, {plata} plata y {oro} oro \n {puntos} puntos \n {enemigos} enemigos amarillo\n")

    rondas = int(tickets/15)
    enemigos = 3 * rondas
    bronce = int(300 * 1.1) * rondas
    puntos = int(45 * 1.05) * rondas
    print(f"Fase 8 {rondas} rondas: \n Estatuas: {bronce} bronce \n {puntos} puntos \n {enemigos} enemigos morado\n")

    plata = int(240 * 1.2) * rondas
    print(f"Fase 9 {rondas} rondas: \n Estatuas: {plata} plata \n {puntos} puntos \n {enemigos} enemigos rojo\n")

    oro = 200 * rondas
    print(f"Fase 10 {rondas} rondas: \n Estatuas: {oro} oro \n {puntos} puntos \n {enemigos} enemigos amarillo\n")
```

```python
def actu_valor_especifico(df:pd.DataFrame,puntos=0, bronce=0, plata=0, oro=0, rojo=0, morado=0, amarillo=0, s_bombardero=0, s_mago=0):
    df.Puntos_Actuales["puntos"] += puntos
    df.Puntos_Actuales["bronce"] += bronce
    df.Puntos_Actuales["plata"] += plata
    df.Puntos_Actuales["oro"] += oro
    df.Puntos_Actuales["morado"] += morado
    df.Puntos_Actuales["rojo"] += rojo
    df.Puntos_Actuales["amarillo"] += amarillo
    df.Puntos_Actuales["s_bombardero"] += s_bombardero
    df.Puntos_Actuales["s_mago"] += s_mago
    
    
    return df
    
```

```python
def actualizar_puntos(fase, rondas, df : pd.DataFrame, path = "puntos.json"):
    bonus_pun = 1.05
    bonus_bro = 1.1
    bonus_pla = 1.2
    bonus_oro = 1
    continuar = True
    try:
        fase = int(fase)
        rondas = int(rondas)
    except:
        continuar=False
        print("La fase y/o las veces completadas no son numeros enteros")
    
    if continuar:
        if rondas > 10:
            print("No se pueden realizar tantas rondas")
            return df
        else:
            enemigos = 3 * rondas
            puntos = int(30 * bonus_pun) * rondas 
            bronce = 0
            plata = 0
            oro = 0
            tipo = "morado"
            match fase:
                case 1:
                    bronce = int(60 * bonus_bro) * rondas
                    plata = int(20 * bonus_pla) * rondas
                case 2:
                    plata = int(50 * bonus_pla) * rondas
                    oro = int(20* bonus_oro) * rondas
                    tipo = "rojo"
                case 3:
                    bronce = int(20 * bonus_bro) * rondas
                    oro = int(50* bonus_oro) * rondas
                    tipo = "amarillo"
                case 4:
                    enemigos = 5 * rondas
                    puntos = int(40 * bonus_pun) * rondas
                    bronce = int(160 * bonus_bro) * rondas
                    tipo = "rojo"
                case 5:
                    puntos = int(40 * bonus_pun) * rondas
                    plata = int(130 * bonus_pla) * rondas
                case 6:
                    puntos = int(40 * bonus_pun) * rondas
                    oro = int(110* bonus_oro) * rondas
                    tipo = "rojo"
                case 7:
                    puntos = int(140 * bonus_pun) * rondas
                    bronce = int(20 * bonus_bro) * rondas
                    plata = int(20 * bonus_pla) * rondas
                    oro = int(20* bonus_oro) * rondas
                    tipo = "amarillo"
                case 8:
                    puntos = int(45 * bonus_pun) * rondas
                    bronce = int(300 * bonus_bro) * rondas
                case 9:
                    puntos = int(45 * bonus_pun) * rondas
                    plata = int(240 * bonus_pla) * rondas
                    tipo = "rojo"
                case 10:
                    puntos = int(45 * bonus_pun) * rondas
                    oro = int(200* bonus_oro) * rondas
                    tipo = "amarillo"
                case other:
                    print("Esa fase no existe")
                    continuar = False
            
            if continuar:          
                if tipo == "amarillo":
                    df = actu_valor_especifico(df, puntos, bronce, plata, oro, amarillo=enemigos)
                elif tipo == "rojo":
                    df = actu_valor_especifico(df, puntos, bronce, plata, oro, rojo=enemigos)
                elif tipo == "morado":
                    df = actu_valor_especifico(df, puntos, bronce, plata, oro, morado=enemigos)
                    
                df["Puntos_Restantes"] = df.Puntos_Maximos - df.Puntos_Actuales
                df.to_json(path)  
            
    return df
                    
```

```python
df = read_df()
decision = input("Que quieres hacer: 1 = comprobar puntos, 2 = aplicar puntos, otros = salir")
try: 
    decision = int(decision)
except:
    pass

match decision:
    case 1:
        tickets = input("Tickets disponibles")
        try:  
            tickets = int(tickets)
            if 180 >= tickets > 0:
                print_info(tickets, read_df())
            else:
                print("No es una cantidad permitida")
        except: 
            print("No son numeros enteros")
                 
    case 2: 
        fase, veces = input("Que fase has completado y cuantas rondas, separado por un espacio").split(" ")
        actualizar_puntos(fase, veces, read_df(), "puntos.json")
    
    case other:
        print("Adios")
```
